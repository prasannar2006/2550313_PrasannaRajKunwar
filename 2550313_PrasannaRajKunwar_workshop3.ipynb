Exploratory Data Analysis Practice Excercise 1

```python
# Fill missing Age values with median age
median_age = df['Age'].median()
df['Age'] = df['Age'].fillna(median_age)
```

```python
#to find which passenger had the highest Fare paid relative to their age
df['fare_per_year'] = df['Fare']/df['Age']
```

```python
high_fare_age = df[df['fare_per_year'] > 5]
high_fare_age_srt = high_fare_age.sort_values(by='fare_per_year', ascending=False)
```

```python
result = high_fare_age_srt[['Name', 'fare_per_year']]
print(result.head())
```

Exploratory Data Analysis with Group-by Method Practice Exercise:

```python
total_fare = df['Fare'].sum()
print(total_fare)
```

```python
fare_first_class = df[df['Pclass'] == 1]['Fare'].sum()
```

```python
fare_second_class = df[df['Pclass'] == 2]['Fare'].sum()
fare_third_class = df[df['Pclass'] == 3]['Fare'].sum()
```

```python
fare_totals = [fare_first_class, fare_second_class, fare_third_class]
```

```python
fare_proportions = [(fare / total_fare) * 100 for fare in fare_totals]

fare_revenue_by_class = pd.DataFrame({
    'Pclass': ['First Class', 'Second Class', 'Third Class'],
    'Fare Revenue (%)': fare_proportions
})

print(fare_revenue_by_class)
```

QUESTION 2

```python
def categorize_age(age):
    if age < 18:
        return 'child'
    elif age < 65:
        return 'adult'
    else:
        return 'senior'

df['age_group'] = df['Age'].apply(categorize_age)
```

```python
total_passengers = len(df)
age_group_counts = df['age_group'].value_counts()
```

```python
age_group_proportions = (age_group_counts / total_passengers) * 100
```

```python
age_group_summary = pd.DataFrame({
    'Age Group': age_group_proportions.index,
    'Passenger Percentage (%)': age_group_proportions.values
})

print(age_group_summary)
```
